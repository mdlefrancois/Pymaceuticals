```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem
import os

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_df = pd.read_csv(mouse_drug_data_to_load, low_memory=False)
clinical_trial_df = pd.read_csv(clinical_trial_data_to_load, low_memory=False)
#clinical_trial_df.head()

# Combine the data into a single dataset
Pharma_combined_df = pd.merge(mouse_drug_df, clinical_trial_df, how='outer', on='Mouse ID')

#Rearange the column output ["Mouse ID", "Timepoint", "Tumor Volume (mm3)", "Metastatic Sites", "Drug"]
Pharma_df_ColOrder = [0,2,3,4,1]
Pharma_df = Pharma_combined_df[Pharma_combined_df.columns[Pharma_df_ColOrder]]

Pharma_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
# Convert to DataFrame
#Note, a pandas series is based on [] brackets -- data.groupby('column title')['column title']
# a pandas data frame is based on [[]] brackets -- ('column title')[['column title']]
#This line of code creates a variable for tumor information and calculates the mean 
Mean_Tumor_Volume = Pharma_df.groupby(['Drug','Timepoint'])[['Tumor Volume (mm3)']].mean()

# Preview DataFrame
Mean_Tumor_Volume.head(12)
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
# Convert to DataFrame
Tumor_Standard_Error = Pharma_df.groupby(['Drug','Timepoint'])[['Tumor Volume (mm3)']].sem()

# Preview DataFrame
Tumor_Standard_Error.head(12)
```

```python
# Minor Data Munging to Re-Format the Data Frames
organized_df = Mean_Tumor_Volume.unstack(level = 0)
organized_df.columns = organized_df.columns.get_level_values(1)

# Preview that Reformatting worked
organized_df
```

```python
# Set up drugs for plotting
#Capomulin
CapoMulin_df = Pharma_df.loc[Pharma_df["Drug"]=="Capomulin", :]
CapoMulin_mean = CapoMulin_df.groupby("Timepoint")["Tumor Volume (mm3)"].mean()
CapoMulin_err = CapoMulin_df.groupby("Timepoint")["Tumor Volume (mm3)"].sem()

#Infubinol
Infubinol_df = Pharma_df.loc[Pharma_df["Drug"]=="Infubinol", :]
Infubinol_mean = Infubinol_df.groupby("Timepoint")["Tumor Volume (mm3)"].mean()
Infubinol_err = Infubinol_df.groupby("Timepoint")["Tumor Volume (mm3)"].sem()

#Ketapril
Ketapril_df = Pharma_df.loc[Pharma_df["Drug"]=="Ketapril", :]
Ketapril_mean = Ketapril_df.groupby("Timepoint")["Tumor Volume (mm3)"].mean()
Ketapril_err = Ketapril_df.groupby("Timepoint")["Tumor Volume (mm3)"].sem()

#Placebo
Placebo_df = Pharma_df.loc[Pharma_df["Drug"]=="Placebo", :]
Placebo_mean = Placebo_df.groupby("Timepoint")["Tumor Volume (mm3)"].mean()
Placebo_err = Placebo_df.groupby("Timepoint")["Tumor Volume (mm3)"].sem()

```

```python
# Generate plots (with Error Bars)
#Using np.arrange to generate a sequence of times on the x axis
#error bars will be the standard error 
x_axis = np.arange(0, 50, 5)

plt.figure(figsize=(12,8))
plt.errorbar(x_axis, CapoMulin_mean, yerr=CapoMulin_err, marker ='o', linestyle='--', alpha = 0.75, label="Capomulin")
plt.errorbar(x_axis, Infubinol_mean, yerr=Infubinol_err, marker ='^', linestyle='--', alpha = 0.75, label="Infubinol")
plt.errorbar(x_axis, Ketapril_mean, yerr=Ketapril_err, marker ='s', linestyle='--', alpha = 0.75, label="Ketapril")
plt.errorbar(x_axis, Placebo_mean, yerr=Placebo_err, marker ='p', linestyle='--', alpha = 0.75, label="Placebo")

plt.suptitle("Mean Tumor Volume Over Time", fontsize=16, fontweight="bold")
plt.xlabel("Time (days)")
plt.ylabel("Mean Tumor Volume (mm3)")
plt.legend(loc="best")
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
# Convert to DataFrame
Mean_Met_Sites = Pharma_df.groupby(['Drug','Timepoint'])[['Metastatic Sites']].mean()

# Preview DataFrame
Mean_Met_Sites.head(12)
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
# Convert to DataFrame
Mean_Met_Sites_err = Pharma_df.groupby(['Drug','Timepoint'])[['Metastatic Sites']].sem()

# Preview DataFrame
Mean_Met_Sites_err.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames - show metastatic treatment per drug
MeanMetSites_df = Mean_Met_Sites.unstack(level = 0)
MeanMetSites_df.columns = MeanMetSites_df.columns.get_level_values(1)

# Preview that Reformatting worked
MeanMetSites_df
```

```python
# Generate the Plot (with Error Bars)
# Set up drugs for plotting
#Capomulin
CapoMulin_df = Pharma_df.loc[Pharma_df["Drug"]=="Capomulin", :]
CapoMulin_mean = CapoMulin_df.groupby("Timepoint")["Metastatic Sites"].mean()
CapoMulin_err = CapoMulin_df.groupby("Timepoint")["Metastatic Sites"].sem()

#Infubinol
Infubinol_df = Pharma_df.loc[Pharma_df["Drug"]=="Infubinol", :]
Infubinol_mean = Infubinol_df.groupby("Timepoint")["Metastatic Sites"].mean()
Infubinol_err = Infubinol_df.groupby("Timepoint")["Metastatic Sites"].sem()

#Ketapril
Ketapril_df = Pharma_df.loc[Pharma_df["Drug"]=="Ketapril", :]
Ketapril_mean = Ketapril_df.groupby("Timepoint")["Metastatic Sites"].mean()
Ketapril_err = Ketapril_df.groupby("Timepoint")["Metastatic Sites"].sem()

#Placebo
Placebo_df = Pharma_df.loc[Pharma_df["Drug"]=="Placebo", :]
Placebo_mean = Placebo_df.groupby("Timepoint")["Metastatic Sites"].mean()
Placebo_err = Placebo_df.groupby("Timepoint")["Metastatic Sites"].sem()

# Show the Figure
x_axis = np.arange(0, 50, 5)

plt.figure(figsize=(12,8))
plt.errorbar(x_axis, CapoMulin_mean, yerr=CapoMulin_err, marker ='o', linestyle='--', alpha = 0.75, label="Capomulin")
plt.errorbar(x_axis, Infubinol_mean, yerr=Infubinol_err, marker ='^', linestyle='--', alpha = 0.75, label="Infubinol")
plt.errorbar(x_axis, Ketapril_mean, yerr=Ketapril_err, marker ='s', linestyle='--', alpha = 0.75, label="Ketapril")
plt.errorbar(x_axis, Placebo_mean, yerr=Placebo_err, marker ='p', linestyle='--', alpha = 0.75, label="Placebo")

plt.suptitle("Metastatic Spread During Treatment", fontsize=16, fontweight="bold")
plt.xlabel("Time (days)")
plt.ylabel("Metastatic Sites")
plt.legend(loc="best")
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
# Convert to DataFrame
Mouse_count = Pharma_df.groupby(['Drug','Timepoint'])[['Mouse ID']].count()

# Preview DataFrame
Mouse_count.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
MouseCount_df = Mouse_count.unstack(level = 0)
MouseCount_df.columns = MouseCount_df.columns.get_level_values(1)

# Preview the Data Frame
MouseCount_df
```

```python
#---------------Calculate the mouse % survival rate for each drug----
#---------------Capomulin--------------------------------------------
# Find mouse survival % rate per drug
CapoMulin_df = Pharma_df.loc[Pharma_df["Drug"]=="Capomulin", :]
CapoMulin_mouse = CapoMulin_df.groupby("Timepoint").count().iloc[:,0]

# Capomulin survival rate
CapoMulin_mouse_r = []
for k in list(CapoMulin_mouse):
    CapoMulin_mouse_r.append((k / CapoMulin_mouse[0])*100)
CapoMulin_mouse

#---------------Infubinol--------------------------------------------
# Find mouse survival % rate per drug
Infubinol_df = Pharma_df.loc[Pharma_df["Drug"]=="Infubinol", :]
Infubinol_mouse = Infubinol_df.groupby("Timepoint").count().iloc[:,0]

#survival rate
Infubinol_mouse_r = []
for k in list(Infubinol_mouse):
    Infubinol_mouse_r.append((k / Infubinol_mouse[0])*100)

#---------------Ketapril--------------------------------------------
# Find mouse survival % rate per drug
Ketapril_df = Pharma_df.loc[Pharma_df["Drug"]=="Ketapril", :]
Ketapril_mouse = Ketapril_df.groupby("Timepoint").count().iloc[:,0]

#survival rate
Ketapril_mouse_r = []
for k in list(Ketapril_mouse):
    Ketapril_mouse_r.append((k / Ketapril_mouse[0])*100)

#---------------Placebo--------------------------------------------
# Find mouse survival % rate per drug
Placebo_df = Pharma_df.loc[Pharma_df["Drug"]=="Placebo", :]
Placebo_mouse = Placebo_df.groupby("Timepoint").count().iloc[:,0]

#survival rate
Placebo_mouse_r = []
for k in list(Placebo_mouse):
    Placebo_mouse_r.append((k / Placebo_mouse[0])*100)

# Show the Figure
x_axis = np.arange(0, 50, 5)

plt.figure(figsize=(12,8))
plt.errorbar(x_axis, CapoMulin_mouse_r, marker ='o', linestyle='--', alpha = 0.75, label="Capomulin")
plt.errorbar(x_axis, Infubinol_mouse_r, marker ='^', linestyle='--', alpha = 0.75, label="Infubinol")
plt.errorbar(x_axis, Ketapril_mouse_r, marker ='s', linestyle='--', alpha = 0.75, label="Ketapril")
plt.errorbar(x_axis, Placebo_mouse_r, marker ='p', linestyle='--', alpha = 0.75, label="Placebo")

plt.suptitle("Percent Mice Survival Rate", fontsize=16, fontweight="bold")
plt.xlabel("Time (days)")
plt.ylabel("Mice Survival Rate %")
plt.legend(loc="best")
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug

#This snippet works as well using loops....
#avg_volume = pd.DataFrame(Pharma_combined_df.groupby(['Drug','Timepoint']).mean()['Tumor Volume (mm3)'])
#tumor_change = []
#for x in range(10):
#        tumor_change.append((avg_volume.iloc[10*(x+1)-1, :] - avg_volume.iloc[10*(x), :])/avg_volume.iloc[10*(x), :] * 100)
#tumor_change

# pull organized data frame from above, use the last and the first in the list for calculation
tumor_change = (organized_df.iloc[-1, :] - organized_df.iloc[0, :])/organized_df.iloc[0, :] * 100
tumor_change
```

```python
# Store all Relevant Percent Changes into a Tuple for plotting
PercentChange = (tumor_change["Capomulin"],tumor_change['Infubinol'],tumor_change['Ketapril'],tumor_change['Placebo'])

# Set x axis and tick locations
fig, ax = plt.subplots(figsize=(7,5))
index = np.arange(len(PercentChange))

# Splice the data between passing and failing drugs
# Create a list indicating where to write x labels and set figure size to adjust for space
#Matplotlib -- Axes.bar(x, height, width=0.8, bottom=None, *, align='center', data=None, **kwargs)
#First item in list is the only negative, [0]
#Other items are all positive [1:]
DrugPass = ax.bar(index[0], PercentChange[0], width = 0.8, bottom=None, color = 'green')
DrugFail = ax.bar(index[1:], PercentChange[1:], width = 0.8, bottom=None, color = 'red')

# Orient widths. Add labels, tick marks, etc. 
#ax.set_xticks(index) #creates an list of numbers 0 - 3 based on PercentChange variable above
plt.xticks(index, PercentChange, rotation="vertical")
ax.set_xticklabels(('Capomulin','Infubinol','Ketapril','Placebo'))
ax.grid(True)

# # set title and axis labels
plt.title("Tumor Change Over 45-Day Treatment")
#plt.xlabel("Drug")
plt.ylabel("% Tumor Volume Change")
plt.xlim(-0.5, len(PercentChange)-0.5)

# Use functions to label the percentages of changes
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x()+rect.get_width()/2., height, '%d'%int(height),
                ha='center', va='bottom', color ='black')
        
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x()+rect.get_width()/2., height, '%d'%int(height),
                ha='center', va='bottom', color ='black')
               
#Call the label function
autolabel(DrugPass)
autolabel(DrugFail)
plt.show()
```
